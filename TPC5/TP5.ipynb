### TPC5: Aplicação para Gerir um Cinema

Suponha que está a desenvolver uma aplicacão para gestão de um conjunto de salas de cinema de um centro comercial. 
Nesse centro comercial existem algumas salas de cinema (que poderão estar a exibir filmes ou não), cada sala tem uma determinada 
lotação, uma lista com a referência dos bilhetes vendidos (lugares ocupados; cada lugar é identificado por um número inteiro), e cada sala tem um filme associado.

Considera a seguinte sugestão para o modelo dos cinemas:
```
Cinema = [Sala]
Sala = [nlugares, Vendidos, filme]
nlugares = Int
filme = String 
Vendidos = [Int]
```
  
Que poderá ser usado num programa da seguinte forma:
```
sala1 = (150, [], "Twilight")
sala2 = (200, [], "Hannibal")
cinema1 = []
...
cinema1 = inserirSala(cinema1,sala1)
cinema1 = inserirSala(cinema1,sala2)
...
listar(cinema1)
...

if(disponivel(cinema1, "Twilight", 17 )):
  cinema1 = vendebilhete(cinema1, "Twilight", 17 )
...
listardisponibilidades(cinema1)
...
```

Especifique as funções utilizadas no exemplo:

1. `listar( cinema )` - que lista no monitor todos os filmes que estão em exibição nas salas do cinema passado como argumento;
2. `disponivel( cinema, filme, lugar )` - que dá como resultado **False** se o lugar lugar já estiver ocupado na sala onde o filme está a ser exibido e dará como resultado **True** se o inverso acontecer;
3. `vendebilhete( cinema, filme, lugar )` - que dá como resultado um novo cinema resultante de acrescentar o lugar à lista dos lugares ocupados, na sala onde está a ser exibido o filme;
4. `listardisponibilidades( cinema )` - que, para um dado cinema, lista no monitor para cada sala, o filme que está a ser exibido e o total de lugares disponíveis nessa sala (número de lugares na sala menos o número de lugares ocupados);
5. `inserirSala( cinema, sala )` - que acrescenta uma sala nova a um cinema (devendo verificar se a sala já existe);
6. Acrescente todas as outras funcionalidades que achar necessárias;
7. À semelhança do exercício 3, construa uma aplicação com um menu de interface para as operações.


````python
#Opção 1 - listar no monitor todos os filmes em exibição nas salas do cinema

def inserirSala(cinema, sala):
    cinema.append(sala)
    return cinema

def listarCinema (cinema):
    for sala in cinema:
        nlugares, vendidos, filmes = sala

        print (f"Sala {cinema.index(sala)+1}: {filmes} filmes disponíveis")

#Opção 2 - Retorna Falso se o lugar já estiver ocupado na sala onde o filme está a ser exibido. Caso contrário, retorna True

def disponivel(cinema, filme, lugar):
    for sala in cinema:
        nlugares, vendidos, filmes = sala
        if filme == filmes:
            if len(vendidos) + lugar <= nlugares:
                print ("Está ocupado.")
            else:
                print ("Não está ocupado.")
    print("A sala não foi encontrada.")
    return False


#Opção 3 - Vender Bilhete que dá como resultado um novo cinema resultante

def venderBilhete (cinema, filme, lugar):
    for i in range (len(cinema)):
        sala = cinema[i]
        nlugares, vendidos, filmes = sala
        if filme == filmes:
            disponiveis = nlugares - len(vendidos)
            if lugar <= disponiveis:
                cinema[i] = (nlugares, vendidos + [i + 1 for i in range(lugar)], filme)
                print(" A quantidade de bilhetes vendidos para o filmes"+str(filme)+"foram "+str(lugar))
            else:
                print("Não há lugares disponíveis para ver esse filme. ")
    return cinema 


#Opção 4 - Listar disponibilidades de um cinema
def listarDisponibilidades (cinema):
    for sala in cinema:
        nlugares, vendidos, filmes = sala
        disponiveis = nlugares - len(vendidos)
        print("O cinema" + "tem"+ str(disponiveis)+"para ver os filmes "+ str(filmes))
    return disponiveis

#Opção 5 - Inserir uma nova sala

def inserirSala(cinema, sala):
    sala1 = sala[2]
    for sala in cinema:
        if sala == sala1:
            cinema.append(sala1)
        else:
            print ("Essa sala já existe no cinema.")
    return cinema

def menu():
    cinema = []
    while True:
        print("Menu:")
        print("1. Listar filmes em exibição")
        print("2. Verificar disponibilidade de lugar")
        print("3. Vender bilhete")
        print("4. Listar disponibilidades de lugares")
        print("5. Inserir nova sala")
        print("6. Sair")
        opcao = input("Escolha uma opção: ")
        if opcao == "1":
            listarCinema(cinema)
        elif opcao == "2":
            filme = input("Nome do filme: ")
            lugar = int(input("Número do lugar: "))
            if disponivel(cinema, filme, lugar):
                print("O lugar"+ str(lugar)+ "está disponível para o filme" +str(filme)+".")
            else:
                print("O lugar"+ str(lugar)+  "já está ocupado para o filme"+str(filme)+".")
        elif opcao == "3":
            filme = input("Nome do filme: ")
            lugar = int(input("Número do lugar: "))
            cinema = venderBilhete(cinema, filme, lugar)
        elif opcao == "4":
            listarDisponibilidades(cinema)
        elif opcao == "5":
            nlugares = int(input("Número de lugares da nova sala: "))
            filme = input("Nome do filme da nova sala: ")
            sala = (nlugares, [], filme)
            cinema = inserirSala(cinema, sala)
        elif opcao == "6":
            print("Bye, Bye!")
        else:
            print("Opção inválida. Tente novamente.")

menu()
````
